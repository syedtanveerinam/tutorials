```python
#@title Convert ipynb to HTML in Colab
# Upload ipynb
from google.colab import files
f = files.upload()

# Convert ipynb to html
import subprocess
import os
file0 = list(f.keys())[0]
_ = subprocess.run(["pip", "install", "nbconvert"])
# Construct the output HTML filename explicitly, replacing spaces with underscores
html_filename = os.path.splitext(file0)[0].replace(" ", "_") + ".html"
_ = subprocess.run(["jupyter", "nbconvert", file0, "--to", "html", "--output", html_filename])


# download the html
files.download(file0[:-5]+"html")
```
